# 음식 필터링
음식 사진 퀄리티가 너무 높아서 '여행지' 라는 키워드에 온전히 집중하지 못 할 것으로 생각되어 따로 분류, 나중에 음식 추천만 따로 해도 될 듯

```python
from fast_ml.utilities import reduce_memory_usage
from tqdm import tqdm

import matplotlib.pyplot as plt
import platform
if platform.system() == 'Darwin': #맥
        plt.rc('font', family='AppleGothic') 
elif platform.system() == 'Windows': #윈도우
        plt.rc('font', family='Malgun Gothic') 
elif platform.system() == 'Linux': # 구글 코랩
        plt.rc('font', family='Malgun Gothic') 
plt.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시 마이너스 폰트 깨짐 해결

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')

df = pd.read_csv("./dataset/수집데이터.csv")
```

# 1차 필터링 수정본

```python
import re
food_keyword = ['음식','한국음식']
point_list1 = []
for i in range(df.shape[0]):
    try:
        if (df.loc[i,'검색 키워드'].find('음식') == -1):
            point_list1.append([df.loc[i,'관광지'], df.loc[i,'검색 키워드']])
    except:
        pass
    
print(f"1차 음식 필터링 결과 : {df.shape[0]:,} -> {len(point_list1):,}개 ({df.shape[0]-len(point_list1):,}개 감소)")
print("[제거대상]")
df.loc[~df['관광지'].isin(np.array(point_list1)[:,0])]
```

```python
result_df = df.loc[df['관광지'].isin(np.array(point_list1)[:,0])].reset_index(drop=True)
result_df
```

```python
result_df.to_csv("./dataset/음식제거.csv",index=False,encoding='utf-8-sig')
```

# 2차 필터링

```python
import re
food_keyword = ['음식','한국음식']
point = ['서울','부산','대구','인천','광주','대전','울산','세종','경기도','강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주']
point_list2 = []
for i in range(df.shape[0]):
    try:
        keywords = df.loc[i,'검색 키워드'].split(',')
        for fk in food_keyword:
            if (fk in df.loc[i,'검색 키워드'].split(',')):
                flag = True
                for p in point: # 지명이 있으면 pass
                    if (p in df.loc[i,'검색 키워드']):
                        flag = False
                        break
                if (flag):
                    point_list2.append([df.loc[i,'관광지'], df.loc[i,'검색 키워드']])
    except:
        pass
    
print(f"2차 음식 필터링 결과 : {len(point_list2):,}개")
df.loc[df['관광지'].isin(np.array(point_list2)[:,0])]
```

```python
df.loc[df['관광지'].isin(list(set(np.array(point_list1)[:,0]) - set(np.array(point_list2)[:,0])))]
```
