```python
def get_locname(strings):
    locname = ['서울','부산','대구','인천','광주','대전','울산','세종','경기도','강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주']
    return_list = []
    for loc in locname:
        if strings.find(loc) != -1:
            return_list.append(loc)
    if len(return_list) == 0:
        print('not in locname : ' + strings)
```

```python
def get_address(address):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
    params = {'query': address}
    headers = {"Authorization": "KakaoAK fa6b74814e224dac2914246ce09d8e18"}
    places = ''
    try:
        places_json = requests.get(url, params=params, headers=headers).json()['documents']
        places = places_json[0]['address_name'].split(' ')[0]
    except Exception as e:
        print(e)
        print(address)
    return places
```
