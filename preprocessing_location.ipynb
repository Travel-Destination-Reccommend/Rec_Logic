```python
def get_locname(strings):
    locname = ['서울','부산','대구','인천','광주','대전','울산','세종','경기도','강원도',
    '충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주','충북','충남','전북','전남','경북','경남','공주']
    locname_rt = ['서울','부산','대구','인천','광주','대전','울산','세종','경기도','강원도',
    '충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','충청남도']
    for i in range(len(locname)):
        if strings.find(locname[i]) != -1:
            return locname_rt[i]
            break
    return ''
```

```python
loc_list = []
for i in range(df.shape[0]):
    if type(df.loc[i, '촬영장소']) == float: #촬영장소 없으면
        if get_locname(df.loc[i, '검색 키워드']) != '': #검색 키워드 살펴보고
            loc_list.append(get_locname(df.loc[i, '검색 키워드'])) #징소 있으면 append
        else:
            loc_list.append(-1) #그마저도 없으면 -1
    else:
        if get_locname(df.loc[i, '촬영장소']) != '': #촬영장소 있으면 살펴보고
            loc_list.append(get_locname(df.loc[i, '촬영장소'])) #멀쩡한 장소 있으면 append
        else:
            if get_locname(df.loc[i, '검색 키워드']) != '': #없으면 키워드에서 살펴보고
                loc_list.append(get_locname(df.loc[i, '검색 키워드'])) #장소 있으면 append
            else:
                loc_list.append(-1) #그마저도 없으면 -1
```
