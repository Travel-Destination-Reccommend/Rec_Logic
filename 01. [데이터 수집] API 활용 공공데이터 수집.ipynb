```python
import requests
from bs4 import BeautifulSoup

key = # key
MobileApp = "test"
numOfRows = 100000
```

```python
total = [] 

url = f"http://api.visitkorea.or.kr/openapi/service/rest/PhotoGalleryService/galleryList?ServiceKey={key}&arrange=B&MobileOS=ETC&MobileApp={MobileApp}&numOfRows={numOfRows}"
res = requests.get(url)

soup = BeautifulSoup(res.content, 'html.parser')

for item in soup.findAll('item'):
    try:
        id = item.find('galcontentid').text
    except Exception as e:
        id = None
    # try:
    #     c_type = item.find('galcontenttypeid').text
    # except Exception as e:
    #     c_type = None
    try:
        a = item.find('galtitle').text
    except Exception as e:
        a = None
    try:
        b = item.find('galwebimageurl').text
    except Exception as e:
        b = None        
    try:
        c = item.find('galphotographylocation').text
    except:
        c = None
    try:
        d = item.find('galsearchkeyword').text.replace(', ',',')
    except Exception as e:
        d = None
    total.append([id,a,b,c,d])
```

```python
df = pd.DataFrame(total, columns=['id','관광지','url','촬영장소','검색 키워드'])
df
```

```python
df.to_csv("./dataset/수집데이터.csv",index=False,encoding='utf-8-sig')
```
